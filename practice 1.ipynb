# List countries or regions with increasing/decreasing trends in cases over a defined period.


```python
covid19_clean['Date'].unique()
```

```python
start_date= '2020-04-12'
end_date='2020-05-12'
covid19_filter=covid19_clean.dropna()
covid19_filter['Date']=covid19_filter['Date'].astype('datetime64[ns]')
```

```python
Filtered= covid19_filter[(covid19_filter['Date']<=start_date) | (covid19_filter['Date']>=end_date)]
Filtered
```

```python
Grouped_Filtered= Filtered.groupby(['Country/Region', 'Date'])['Confirmed'].sum().reset_index()
Grouped_Filtered
```

```python
start_case= Grouped_Filtered[Grouped_Filtered['Date']==start_date][['Country/Region', 'Confirmed']].set_index('Country/Region')

end_case= Grouped_Filtered[Grouped_Filtered['Date']==end_date][['Country/Region', 'Confirmed']].set_index('Country/Region')

diff= start_case.join(end_case,lsuffix='_start', rsuffix='_end')
diff['trend']= diff['Confirmed_end'] - diff['Confirmed_start']
diff['trend'] =diff['trend'].apply(lambda x: 'Decreasing' if x<0 else 'Increasing'  if x>0 else 'No chage')
```

```python
diff=diff['trend']
```

# Plot line charts of total confirmed, deaths, recovered, and active cases for a chosen country or globally.


```python
line=full_grouped
country='Canada'
line=full_grouped[full_grouped[	'Country/Region']==country][['Date',	'Country/Region',	'Confirmed',	'Deaths',	'Recovered',	'Active']]
line['Date'] = pd.to_datetime(line['Date'])
line.set_index('Date', inplace=True)

plt.figure(figsize=(10,6))

plt.plot(line.index, line['Recovered'], label='Total Recovered')
plt.plot(line.index, line['Deaths'], label='Deaths')
plt.plot(line.index, line['Active'], label='Active Cases')

plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Trends: Recovered, Deaths, and Active Cases Over Time')
```

# Aggregate daily case numbers to weekly or monthly totals for a particular country or continent.


```python
full_grouped['Date']=full_grouped['Date'].astype('datetime64[ns]')
full_grouped['Date']=full_grouped['Date'].dt.month
```

```python
monthly_totals = full_grouped.groupby('Date')[['New cases']].sum().reset_index()
```

```python
monthly_totals.rename(columns={'Date': 'Month'},inplace=True)
```
